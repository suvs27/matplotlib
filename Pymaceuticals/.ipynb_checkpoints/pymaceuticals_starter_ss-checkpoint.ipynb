```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mdata = pd.read_csv(mouse_drug_data_to_load)
#mdata.head()
ctd = pd.read_csv(clinical_trial_data_to_load)
#ddata.head()

# Combine the data into a single dataset
print("mdata diminisions: {}".format(mdata.shape))
print("ctd diminisions: {}".format(ctd.shape))

# Display the data table for preview
mergetable = pd.merge(ctd, mdata, on="Mouse ID", how="left")
print("mergetable diminisions: {}".format(mergetable.shape))
mergetable.head()



```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mergetable.columns
trt = mergetable[["Drug","Timepoint","Tumor Volume (mm3)"]]
trt.head()
grouped_trt = trt.groupby(["Drug", "Timepoint"])

mean_tumor_volume = grouped_trt["Tumor Volume (mm3)"].mean()
mean_tumor_volume


# Convert to DataFrame
trt2 = pd.DataFrame({"Tumor Volume (mm3)":mean_tumor_volume})
trt2 = trt2.reset_index()


# Preview DataFrame
trt2
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volume = grouped_trt["Tumor Volume (mm3)"].sem()
sem_tumor_volume

# Convert to DataFrame
trt3 = pd.DataFrame({"Tumor Volume (mm3)":sem_tumor_volume})


# Preview DataFrame
trt3 = trt3.reset_index()
trt3.head()

```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = trt4
percent_change45 = ((((percent_change.loc[45,:]) - 45)/45)*100)

# Display the data to confirm
percent_change45

```

```python
# Store all Relevant Percent Changes into a Tuple
fig, ax = plt.subplots()
#percent_tuple = tuple(percent_change45)
#percent_tuple

drug_names = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
drug_names

drug_name_below_zero = []
percent_below_zero = []

drug_name_above_zero = []
percent_above_zero = []

# Create a bar chart based upon the above data
# Splice the data between passing and failing drugs
# Call functions to implement the function calls
for drug in drug_names:
    if percent_change45[drug] <0:
        drug_name_below_zero.append(drug)
        percent_below_zero.append(percent_change45[drug])
    else:
        drug_name_above_zero.append(drug)
        percent_above_zero.append(percent_change45[drug])

fig.set_figheight(20)
fig.set_figwidth(20)
ax.bar(drug_name_below_zero, percent_below_zero, color = "green", align="center")
ax.bar(drug_name_above_zero, percent_above_zero, color = "red", align="center")
#ax1.bar(drug_name_below_zero, percent_below_zero, color = "blue", align="center")
#ax2.bar(drug_name_above_zero, percent_above_zero, color = "pink", align="center")
#ax3.bar(drug_name_above_zero, percent_above_zero, color = "pink", align="center")
#ax3.bar(drug_name_below_zero, percent_below_zero, color = "yellow", align="center")

# Give the chart a title, x label, and y label
# Orient widths. Add labels, tick marks, etc. 
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("%Tumor Volume Change")
plt.ylim(-30, 50+20)
plt.grid(True)

# Use functions to label the percentages of changes
def labelnegativepercent(i, dr):
    a = round((percent_change45[dr]))
    ax.text(i,a+1,a)
    
def labelpositivepercent(i, dr):
    a = round((percent_change45[dr]))
    ax.text(i, a-2,a)

# Call functions to implement the function calls
j=0
for drug in drug_name_below_zero:
    labelnegativepercent(j,drug)
    j=j+1

i=1
for drug in drug_name_above_zero:
    labelpositivepercent(i,drug)
    i=i+1
    

# Save the Figure
fig.savefig("Tumor Change Over 45 Day Treatment")

# Show the Figure
fig.show()
```

```python
# Three Observations About the Data 
# (1) Over the course of 45 days, Capomulin, Infubinol, Katapril, and Placebo was administered to 4 different sets containing 25 mice each.
# (2) Out of these four drugs, the set of mice treated with Capomulin showed the greatest decrease (approximately -19%) in the percent of tumor volume change over 45 days.  
# (3) Similarly, at the end of the 45 days, 84% of the mice treated with Capomulin survived.
```
